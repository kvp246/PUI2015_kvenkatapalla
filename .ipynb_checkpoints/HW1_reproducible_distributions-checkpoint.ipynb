```python
# Importing required libraries
%pylab inline
import numpy as np
import matplotlib.pylab as plt

# Generating random 2-D objects from gaussian distributions
np.random.seed(123) # Setting up seed for random number generation to enable reproducibility
data1 = np.random.randn(100, 2) # Generating 100 2-D data points in a standard gaussian distribution
data2 = 2 * np.random.randn(50, 2) + [1,1] # Generating 50 2-D data points in a gaussian distribution with different parameters
data1 = data1.T # Transposing the initial output of 100 random 2-D data points
data2 = data2.T # Transposing the initial output of 50 random 2-D data points

# Genarating scatter plot using the two sample outputs as a single distribution
plt.scatter(data1[0],data1[1])
plt.scatter(data2[0],data2[1])
```

```python
# Generating scatter plot using the two sample outpus with color coding, title, and labels
plt.scatter(data1[0],data1[1],c='green',label=r'$\mu = 0\/\sigma = 1$')
plt.scatter(data2[0],data2[1],c='yellow',label=r'$\mu = 1\/\sigma = 2$')
pylab.legend(loc='upper right')
plt.title('Scatter plot of 2 random samples')
plt.xlabel('Range')
plt.ylabel('Random Value')
```
